# Data Imputation

* handling missing values 

* Missing values are replaced by 
    * using mean()
    * using Median()
    * using mode()
    * constant()

```python
from  sklearn.impute import SimpleImputer
```

```python
di = {
    "Name" : [np.nan,"madhan","trinadh","harsha"],
    "age" : [23,22,21,20],
    "Gender" : ["M","M","f","f"]
}
df = pd.DataFrame(di)
df
```

```python
smp.fit(df)
```

```python
smp.transform(df)
```

```python
di = {
    "Name" : [12,12,54,np.nan],
    "age" : [23,22,np.nan,20],
    "Gender" : [12,0,1,np.nan]
}
df = pd.DataFrame(di)
df
```

```python
smp = SimpleImputer(strategy= "median")
smp
```

```python
smp.fit(df)
```

```python
smp.transform(df)
```

```python
df.median() 
```

```python
smp = SimpleImputer(strategy= "mean")
smp.fit_transform(df)
```

```python
df.mean()
```

```python
smp = SimpleImputer(strategy= "most_frequent") # mode
smp.fit_transform(df)
```

```python
df.mode()
```

```python
smp = SimpleImputer(strategy= "constant", fill_value = -1) 
smp.fit_transform(df)
```
